# Visualization

```python
hero_columns = X.columns[2:115]
heroes = hero_columns.tolist()

def get_selected_heroes(row):
    team1_heroes = [hero for hero in heroes if row[hero] == 1]
    team2_heroes = [hero for hero in heroes if row[hero] == -1]
    return team1_heroes, team2_heroes

TOTAL = 50000

visinfo = pd.DataFrame(X[:TOTAL])

visinfo['team1_heroes'], visinfo['team2_heroes'] = zip(*visinfo.apply(get_selected_heroes, axis=1))

# Function to generate all possible hero pairs within a team
def generate_pairs(hero_list):
    return list(combinations(sorted(hero_list), 2))

# Apply to both teams
visinfo['team1_pairs'] = visinfo['team1_heroes'].apply(generate_pairs)
visinfo['team2_pairs'] = visinfo['team2_heroes'].apply(generate_pairs)

# Combine all pairs from both teams
all_pairs = visinfo['team1_pairs'].tolist() + visinfo['team2_pairs'].tolist()

# Flatten the list of lists
from itertools import chain
all_pairs = list(chain.from_iterable(all_pairs))

def generate_matchups(team1, team2):
    return list(product(sorted(team1), sorted(team2)))

pair_counts = Counter(all_pairs)

# Initialize dictionaries to store counts
pair_win_counts = Counter()
pair_total_counts = Counter()

for index, row in visinfo.iterrows():
    # Check team1
    team1_pairs = row['team1_pairs']
    for pair in team1_pairs:
        pair_total_counts[pair] += 1
        if y['win'][index] == 1:
            pair_win_counts[pair] += 1
    # Check team2
    team2_pairs = row['team2_pairs']
    for pair in team2_pairs:
        pair_total_counts[pair] += 1
        if y['win'][index] == -1:
            pair_win_counts[pair] += 1

# Calculate success rates
pair_success_rate = {pair: pair_win_counts[pair] / pair_total_counts[pair] 
                    for pair in pair_total_counts if pair_total_counts[pair] >= 0}  # Consider pairs with at least 10 matches

visinfo['matchups'] = visinfo.apply(lambda row: generate_matchups(row['team1_heroes'], row['team2_heroes']), axis=1)
all_matchups = list(chain.from_iterable(visinfo['matchups']))
matchup_counts = Counter(all_matchups)

# Initialize dictionaries to store counts
matchup_win_counts = Counter()
matchup_total_counts = Counter()

for index, row in visinfo.iterrows():
    team1 = row['team1_heroes']
    team2 = row['team2_heroes']
    matchups = row['matchups']
    outcome = y['win'][index]  # 1 if Team1 won, -1 if Team2 won
    
    for matchup in matchups:
        hero_A, hero_B = matchup
        matchup_total_counts[matchup] += 1
        if outcome == 1:
            # Team1 won; hero_A's team won against hero_B
            matchup_win_counts[matchup] += 1
        elif outcome == -1:
            # Team2 won; hero_B's team won against hero_A
            # For hero_A, it's a loss against hero_B
            # Depending on how you want to represent, you might track this differently
            pass  # Only increment win counts when hero_A's team wins
    
# Calculate success rates
# Success rate of hero_A against hero_B
matchup_success_rate = {pair: matchup_win_counts[pair] / matchup_total_counts[pair]
                        for pair in matchup_total_counts if matchup_total_counts[pair] >= 0} 
```

## Heatmap

Initially, I tried to visualize a network of pairs of heros with colors and edge thicknesses representing things like frequency and success rate, but it wasn't working.

Instead, what I've done is created two heatmaps, one for the success rates for pairs of heroes on the same team (symmetrical), and another for the success rate of a hero matched up against some other on the enemy team (assymetrical).
Dota is a 5v5, so technically the relationships are more complex than just the matchups of two individuals within the game of 10 players, but players do tend to pick for and play towards specific synergies or counters, and so I felt like they are meaningful visualizations even if they cannot capture all of the dimensions involved.

```python
import pandas as pd

pairwise_counts = pair_counts

# Extract unique heroes
heroes = set()
for hero1, hero2 in pairwise_counts.keys():
    heroes.add(hero1)
    heroes.add(hero2)
heroes = list(heroes)  # Sorting for consistency

# Initialize a DataFrame with zeros
heatmap_df = pd.DataFrame(0.0, index=heroes, columns=heroes)

# Populate the DataFrame with pairwise counts
for (hero1, hero2), count in pairwise_counts.items():
    success_rate = pair_success_rate.get((hero1, hero2))
    if success_rate is None:
        success_rate = pair_success_rate.get((hero2, hero1))
    if success_rate is None:
        continue
    heatmap_df.loc[hero1, hero2] = success_rate
    heatmap_df.loc[hero2, hero1] = success_rate

# Optionally, set the diagonal to zero if you don't want self-pairings
for hero in heroes:
    heatmap_df.loc[hero, hero] = 0

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 25))  # Adjust the size as needed

sns.heatmap(heatmap_df, cmap='YlGnBu', robust=False)
#sns.clustermap(heatmap_df, cmap='YlGnBu', figsize=(20, 20), linewidths=.5)

plt.title('Pairwise Hero Selection Heatmap')
plt.xlabel('Heroes')
plt.ylabel('Heroes')
plt.show()
```

Assuming that a 50/50 "random guess" would produce a Brier Score of .25, the model currently does slightly better than random guessing. This is about what is expected for the average experienced Dota 2 player. This is because the outcome of the game is influenced by more than just the draft, which is all of the features we have.

However, based on the calibration curve, there is clearly some room for improvement. There seem to be scenarios on the extremes where we should be relatively confident in the outcome (closer to 80-90%) but where the model remains conservative in its predictions.
